```python
import pandas as pd
```

# 1. Importação dos dados

```python
# URL do conjunto de dados
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

# Definir nomes das colunas
column_names = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",
                "num-of-doors", "body-style", "drive-wheels", "engine-location",
                "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
                "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
                "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
```

```python
# Importar o conjunto de dados
df = pd.read_csv(url, names=column_names, na_values="?")
df.head()
```

# 2. Descrição estatística

```python
df.info()
```

```python
# descrição estatística das colunas numéricas
# contagem de registros, média, desvio padrão, valores mínimo e máximo e os quartis
df.describe()
```

# 3. Transformação dos dados

```python
# forma do data frame
df.shape
```

```python
# contagem de cada coluna excluindo os nulos
df.count()
```

```python
# substituir valores faltantes por 0
df.fillna(0, inplace=True)

df.count()
```

```python
# a coluna 'price' já está como float
df['price'].dtype
```

# 4. Visualização de dados

```python
import matplotlib.pyplot as plt

# Gráfico de dispersão entre 'engine-size' e 'price'
plt.figure(figsize=(10, 6))
plt.scatter(df['engine-size'], df['price'], alpha=0.5)
plt.title('Variação de Preço em relação ao Tamanho do Motor')
plt.xlabel('Tamanho do Motor (engine-size)')
plt.ylabel('Preço (USD)')
plt.grid(True)
plt.show()
```
