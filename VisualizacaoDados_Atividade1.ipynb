```python
import numpy as np
import matplotlib.pyplot as plt 

idades = np.array([25, 30, 35, 40, 45, 50, 55, 60]) 
salarios = np.array([5000, 6000, 5500, 7000, 6500, 7500, 8000, 9000]) 

media_salarios = np.mean(salarios) 
desvio_padrao_salarios = np.std(salarios) 

# Geração do gráfico com Matplotlib 
plt.plot(idades, salarios, marker='o') 
plt.xlabel('Idades') 
plt.ylabel('Salários') 
plt.title('Relação Idades-Salários') 
plt.grid(True) 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar Series para as variáveis Idades e Salários
idades = pd.Series([25, 30, 35, 40, 45, 50, 55, 60], name='idades')
salarios = pd.Series([5000, 6000, 5500, 7000, 6500, 7500, 8000, 9000], name='salarios')

# Juntar as Series em um DataFrame
df = pd.concat([idades, salarios], axis=1)

# Calcular a média ods salários
media_salarios = round(df['salarios'].mean(), 2)

# Calcular o desvio padrão dos salários
desvio_padrao_salarios = round(df['salarios'].std(), 2)

print(f"A Média dos Salários é {media_salarios}. \nO Desvio Padrão dos Salários é: {desvio_padrao_salarios}")
```

```python
# Geração do gráfico com Pandas
df.plot(x='idades', y='salarios', kind='line', marker='o')
plt.xlabel('Idades')
plt.ylabel('Salários')
plt.title('Relação Idades-Salários')
plt.grid(True)
plt.show()
```
